```python
from astropy.io import ascii
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rc

data_lvdt_Nov17 = np.genfromtxt('oldsprings/LVDT_test_Nov2017.dat',comments="#",
                                       dtype=(float,float),names="El,LVDTB")


data_lvdt_May21 = np.genfromtxt('oldsprings/LVDT_test_May21_2021.dat',comments="#",
                                       dtype=(float,float,float,float),names="El,LVDTA,LVDTB,LVDTC")
data_lvdt_June8 = np.genfromtxt('newsprings/LVDT_test_June8.dat',comments="#",
                                       dtype=(float,float,float,float),names="El,LVDTA,LVDTB,LVDTC")

El_May21 = data_lvdt_May21['El']
LVDTB_May21 = data_lvdt_May21['LVDTB']
LVDTA_May21 = data_lvdt_May21['LVDTA']
LVDTC_May21 = data_lvdt_May21['LVDTC']

El_June8 = data_lvdt_June8['El']
LVDTB_June8  = data_lvdt_June8['LVDTB']
LVDTA_June8 = data_lvdt_June8['LVDTA']
LVDTC_June8 = data_lvdt_June8['LVDTC']


El_Nov17 = data_lvdt_Nov17['El']
LVDTB_Nov17 = data_lvdt_Nov17['LVDTB']

LVDTB_diff_May21 = 1000.*(LVDTB_May21 - LVDTB_May21[0])
LVDTA_diff_May21 = 1000.*(LVDTA_May21 - LVDTA_May21[0])
LVDTC_diff_May21 = 1000.*(LVDTC_May21 - LVDTC_May21[0])

LVDTB_diff_June8 = 1000.*(LVDTB_June8 - LVDTB_June8[0])
LVDTA_diff_June8 = 1000.*(LVDTA_June8 - LVDTA_June8[0])
LVDTC_diff_June8 = 1000.*(LVDTC_June8 - LVDTC_June8[0])

LVDTB_diff_Nov17 = 1000.*(LVDTB_Nov17 - LVDTB_Nov17[0])

#LVDTB_diff_May21

#El_May21 = data_lvdt_May21_v2[][0]
#print(El_May21)
```

```python
plt.figure(figsize=(16,8))

#plot LVDT changes with elevation, May 21
plt.plot(El_May21,LVDTB_diff_May21,'bo',label='LVDT B; May 21')
plt.plot(El_June8,LVDTB_diff_June8,'ro',label='June 8; New Springs')

plt.plot(El_May21,LVDTA_diff_May21,'bs',label='LVDT A; May 21')
plt.plot(El_June8,LVDTA_diff_June8,'rs',label='June 8; New Springs')

plt.plot(El_May21,LVDTC_diff_May21,'b*',label='LVDT C; May 21')
plt.plot(El_June8,LVDTC_diff_June8,'r*',label='June 8; New Springs')

plt.xticks(fontsize=22)
plt.yticks(fontsize=22)

plt.gca().set_xlabel('Elevation (deg)',fontsize=22)
plt.gca().set_ylabel('Delta LVDT',fontsize=22)


plt.legend(loc='best',handlelength=2,fontsize=22)

plt.savefig('LVDT_vs_El_all',bbox_inches='tight')

plt.show()
```
